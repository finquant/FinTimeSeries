# **ส่วนที่ 1 : การวิเคราะห์อนุกรมเวลาตัวแปรเดียว (Univariate Time Series Analysis)** <a class="anchor" id="0.2"></a>

# **บทที่ 1. แนวคิดพื้นฐานเกี่ยวกับอนุกรมเวลา** <a class="anchor" id="1"></a>

[สารบัญ](#0.1)

## **1.1 ความหมายของอนุกรมเวลา**
อนุกรมเวลาแบบสุ่มของตัวแปรหนึ่ง เขียนแทนด้วย $X_t$

- ค่าเฉลียของ $X_t$ คือ $\mathbb{E}[X_t]$ = $\mu_t$
- ค่าความแปรปรวนของ $X_t$ เขียนแทนด้วย $var(X_t)=\sigma^2_t$
- ค่าความแปรปรวนร่วมระหว่างช่วงเวลา $t_1$ และ $t_2$ ของ $X_t$ เขียนแทนด้วย $cov(X_{t_1},X_{t_2}) = \gamma_{t_1,t_2}$

ในกรณีที่ $X_t$ ถูกเก็บรวบรวมต่อเนื่องกันเป็นจำนวน T ช่วงเวลา จะพบว่าจำนวนพารามิเตอร์ของอนุกรมเวลาคือ

- $\mu_t$ จำนวน T ตัวได้แก่ $\mu_1, \mu_2, ... , \mu_T$
- $\sigma^2_t$ จำนวน T ตัวได้แก่ $\sigma_1^2, \sigma_2^2, ..., \sigma_T^2$
- $\gamma_{t_1,t_2}$ จำนวน ${ T \choose 2}$ คือ $\frac{T(T-1)}{2}$

รวมพารามิเตอร์ทั้งหมดคือ $T+T+\frac{T(T-1)}{2}=\frac{T(T+3)}{2}$

## **1.2 แนวคิดการพยากรณ์อนุกรมเวลา**

<u>**Assumption</u> :** รูปแบบที่ระบุได้จากข้อมูลยอดในอดีต ต้องเหมือนเดิมหรือไม่เปลี่ยนแปลงในอนาคต

## **1.3 ส่วนประกอบของอนุกรมเวลา**

ประกอบด้วย

1. เส้นแนวโน้ม (Trend): แสดงทิศทาง
2. วัฎจักร (Cycle) : ไม่ตรงตามเวลาปฎิทิน กินเวลาตั้งแต่ 2 ปี ถึง 10 ปี
3. ความผันแปรจากฤดูกาล (Seasonal Variations) : รูปที่เกิดขึ้นในรอบ 1 ปีตามเวลาปฎิทิน
4. ความผันผวนจากเหตุการณ์ไม่ปกติ (Irregular Fluctuations) : เป็นเหตุการณ์ไม่คาดฝัน (Shock) เช่น แผ่นดินไหว สึนามิ ระเบิด การประท้วง

$X_t$ อาจเกิดจาก $ Trend + Cycle + Seasonal + Irregular $ หรือ $ Trend*Cycle*Seasonal*Irregular $ ก็ได้

## **1.4 วิธีการพยากรณ์**
แบ่งเป็นสองวิธี
1. เชิงคุณภาพ (Qualitative Forecasting Method) : ใช้ความเห็นของผู้เชี่ยวชาญ มักใช้ในกรณีที่ไม่สามารถหาข้อมูลได้
2. เชิงปริมาณ (Quantitative Forecasting Method) : ในที่นี้จะใช้วิธีตาม Box-Jenkins โดยมีเงื่อนไขสองข้อคือ
   - ต้องมีความนิ่ง (Stationary)
   - ไม่มีเงื่อนไขของฤดูกาล (No Seasonal Variations)
   
วิธีการของ Box-Jenkins

ขั้นที่ 1 : ระบุแบบจำลอง

    - ดูจากจำนวนพจน์ที่เกิด Autocorrelation
        . ค่า SAC (Sample Autocorrelation Function)
        . ค่า SPAC(Sample Partial Autocorrelation Function)

ขั้นที่ 2 : ประมาณค่าพารามิเตอร์ \
ขั้นที่ 3 : ตรวจสอบแบบจำลอง

    - Error Term ต้อง Uncorrelated ต่อกัน

## **1.5 ค่าคลาดเคลื่อนในการพยากรณ์**
Forecaseting Error $e = X_t - \hat{X_t}$ คือ ความผันผวนจากเหตุการณ์ไม่ปกติ

- $e$ ต้องเป็นตัวแปรสุ่ม แบบจำลองจึงสามารถสะท้อนทั้ง แนวโน้ม วัฎจักร และ ฤดูกาล


```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

NT = 100 # จำนวน Step คือ 100
err  = np.random.normal(0,1,(NT,1))

mpl.rcParams['font.family'] = 'Tahoma'

plt.style.use('seaborn-v0_8-poster')
plt.scatter(range(NT),err)
plt.grid('minor')
plt.axis('tight')
plt.title('ค่าความคลาดเคลื่อนที่มีลักษณะของตัวแปรสุ่ม')
plt.xlabel('ช่วงเวลา')
plt.ylabel('ค่าความคลาดเคลื่อน')
plt.show()
```

## **1.6 การวิเคราะห์ความถูกต้องของผลการพยากรณ์**

ค่า Root Mean Square Error, RMSE
$$RMSE = \sqrt{\frac{\sum_{t=1}^{T}{X_t-\hat{X_t}}}{T}}$$

```python
# คำนวณค่า RMSE
RMSE = np.sqrt(err.mean())
print(RMSE)
```

# **บทที่ 2. ความนิ่งของอนุกรมเวลา ค่า SAC และ ค่า SPAC** <a class="anchor" id="2"></a>

## **2.1 ความนิ่งของข้อมูล (Stationary)**

อนุกรมเวลาของตัวแปร $X_t$ จะมีความนิ่งก็ต่อเมื่อ 

1. ค่าเฉลี่ยของตัวแปร $X$ ในแต่ละช่วงเวลา $t$ มีค่าคงที่ $\mathbb{E}[X_t]$ = $\mu; t=1,2,...,T$ ~ 1 พารามิเตอร์
2. ความแปรปรวนของตัวแปร $X$ ในแต่ละช่วงเวลา $t$ มีค่าคงที่ $var(X_t) = \sigma^2; t=1,2,...,T$ ~ 1 พารามิเตอร์
3. ความแปรปรวนร่วมของตัวแปร $X$ ณ เวลา $t_1$ และ $t_2$ มีค่าคงที่ $cov(X_{t_1},X_{t_2})=\gamma_{\tau}; \tau=t_2-t_1$ ~ T-1 พารามิเตอร์

จากเงื่่อนไข Stationary ทำให้จำนวนพารามิเตอร์ทั้งหมดมีเพียง 1 + 1 + (T-1) = T + 1 ตัว

ตัวอย่างอนุกรมเวลาที่ไม่มีความนิ่ง \
ให้ $Y_t$ คืออนุกรมเวลายอดขายแฮมเบอร์เกอร์ยี่ห้อหนึ่งเป็นรายเดือน จำนวน 120 เดือนตั้งแต่มกราคม 2544 - ธันวาคม 2553 หรือ t = 1,2,...,120

```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import matplotlib as mpl
import matplotlib.pyplot as plt   # data visualization


import pandas as pd 
df = pd.read_excel(r'input/ch2_1.xlsx')
df.head()

mpl.rcParams['font.family'] = 'Tahoma'

plt.style.use('seaborn-v0_8-poster')
plt.plot(df['y'])
plt.title('แสดงยอดขายรายเดือนของแฮมเบอร์เกอร์ยี่ห้อหนึ่ง')
plt.xlabel('เวลา')
plt.ylabel('ยอดขาย (ล้านบาท)')
plt.axis('tight')
plt.grid('minor')
plt.grid('on')
```

จากรูปพบว่าข้อมูลไม่มีความนิ่ง (Stationary) การแก้ไขสามาถใช้ผลต่างลำดับที่ 1 (Fist Difference) มาปรับข้อมูลให้มีความนิ่งได้
$$ \Delta{X_t} = X_t - X_{t-1}; t= 2,3,..,T $$

```python
DY=df.diff().dropna()
plt.style.use('seaborn-v0_8-poster')
plt.plot(DY['y'])
plt.title('ผลต่างลำดับที่ 1 ของยอดขายแฮมเบอร์เกอร์ยี่ห้อหนึ่ง')
plt.xlabel('ลำดับที่')
plt.ylabel('ผลต่างลำดับที่หนึ่งของยอดขาย')

plt.hlines(DY['y'].mean() ,0,DY['y'].count()+1,linestyles='dotted')
plt.legend({'DY','ค่าเฉลี่ย'},loc='best')
plt.axis('tight')
plt.grid('minor')
plt.grid('on')
plt.show()
```

จะได้ว่า fist difference ของ $Y_t$ มีความนิ่ง มีความผันผวนคงที่ และค่าเฉลี่ยเป็นศูนย์

## **2.2 Sample Autocorrelation Function (SAC)**
ใช้ตรวจสอบ Correlation ระหว่าง $X_1, X_2,...,X_T$ 

**Theoretical Autocorrelation Function (TAC)** \
Assume: Stationary , so $var(Z_t)=var(Z_{t+k})), \mathbb{E}Z_t=\mathbb{E}Z_{t+k}$
$$\rho_k =\frac{cov(Z_t,Z_{t+k})}{var(Z_t)}=\frac{\mathbb{E}[Z_t-\mathbb{E}Z_t][Z_{t+k}-\mathbb{E}Z_t]}{\mathbb{E}[Z_t-\mathbb{E}Z_t]}$$
**Sample Autocorrelation Function (SAC)** \
Correlation ระหว่าง time t กับ t-k
$$r_k=\frac{\sum_{t=b}^{n-k}[Z_t-\bar{Z}][Z_{t-k}-\bar{Z}]}{\sum_{t=b}^{n}[Z_t-\bar{Z}]^2} $$ \
**Hypothesis Testing** \
$$H_0: \rho_0=\rho_1=...=\rho_m$$
$$ H_1:  \exists \rho_i \neq 0; i=0,1,..,m $$
ค่าสถิติ Ljung-Box Q \
$$ Q(m)=T(T+2)\sum_{j=1}^{n}\frac{r_j^2}{T-j}$$

```python
# import statsmodels.api as sm
import statsmodels.api as sm
sm.stats.acorr_ljungbox(df['y'], lags=[10], return_df=True)
```

ยืนยันว่ายอดขายแฮมเบอร์เกอร์หรือ $Y_t$ ไม่เป็นอิสระกับช่วงเวลาอื่น และพบปัญหา Autocorrelation อย่างมีนัยสำคัญ

```python
sm.stats.acorr_ljungbox(DY['y'], lags=[10], return_df=True)
```

ส่วน Fist Difference หรือ $DY_t$ ซึ่งเป็นข้อมูลที่มีความนิ่งนั้นก็ยังไม่เป็นอิสระกับช่วงเวลาอื่นเช่นกัน

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Draw Plot
mpl.rcParams['font.family'] = 'Tahoma'

plt.style.use('seaborn-v0_8-poster')
fig, axes = plt.subplots(1,2,figsize=(16,3), dpi= 100)
plot_acf(df['y'].tolist(), lags=50, ax=axes[0],title="ค่า SAC ของยอดขาย $Y_t$")
plot_acf(DY['y'].tolist(), lags=50, ax=axes[1],title="ค่า SAC ของ Fist Difference, $DY_t$")
plt.axis('tight')
plt.grid('minor')
plt.grid('on')
plt.show()
```

## **2.3 Sample Partial Autocorrelation Function (SPAC)**
ในกรณีของ SPAC ณ k ช่วงเวลาที่แล้ว
$$Z_t=\phi_{0,k}+\phi_{1,k}Z_{t-1}+\phi_{2,k}Z_{t-2}+\phi_{3,k}Z_{t-3}+...++\phi_{k,k}Z_{t-k} $$
<center>where $\phi_{k,k}$ คือ ค่า TPAC ระหว่าง $Z_t$ กับ $Z_{t-k}$ หรือ K ช่วงเวลาที่แล้ว.ไม่พิจารณาความสัมพันธ์ของ$Z_t$ กับ $Z_{t-1}$ ถึง $Z_{t-k}$ </center>


**การทดสอบสมมุติฐาน :** ดูเส้นประสองเส้นที่อยู่ตรงกราฟของ PAC


จากแผนภูมิพบว่าค่า SAC ลดลงอย่าง exponential ส่วนค่า SPAC นั้นมีความสำคัญถึง Lag 1 ดังนั้นแบบจำลองที่เหมาะสมคือ AR(1)

## **2.4 การกำจัดความผันแปรจากฤดูกาล (Removal Seasonal Variations)**

2. ไม่มีความผันแปรจากฤดูกาล

# **บทที่ 3. วิธีการของ Box-Jenkins: การระบุแบบจำลอง** <a class="anchor" id="3"></a>

[สารบัญ](#0.1)


## **3.1 ตัวรบกวนขาว (White Noise)**
ให้ $\epsilon_t$ เป็น Gaussian white noise จะได้ว่า $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$

ไม่สามารถปฎิเสธสมมุติฐานหลักของ SAC คือ $H_0 : \rho_1=\rho_2=...=\rho_m=0$ \
ไม่สามารถปฎิเสธสมมุติฐานหลักของ SPAC คือ $H_0 : \phi_{k,k}=0;k \neq 0$

จึงเรียก $\epsilon_t$ ว่าไม่มีความจำ (no memory)


## **3.2 แบบจำลองของ Box-Jenkins**
### **3.2.1 แบบจำลอง Autoregressive (AR)** 
**(1) แบบจำลอง Autoregressive ลำดับที่ 1** \
แบบจำลอง AR(1) อยู่ในรูป
$$X_t=\alpha_0+\alpha_1X_{t-1}+\epsilon_t$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise </center>
    
จาก **Assumption : Stationary** จะได้ว่า
$$\mu=\frac{\alpha_0}{1-\alpha_1}$$
<center>$\gamma_0=\frac{\sigma^2}{1-\alpha_1^2}$ ซึ่งมีนัยว่า $|\alpha_1|<1$ </center>
$$\rho_k=\alpha_1^k$$
    
- การที่ $|\alpha_1|<1$ ทำให้ค่า SAC ของ AR(1) จะลดลงแบบ Exponential
    
- การที่ AR(1) มีเพียง Lag 1 เท่านั้นค่า SPAC จึงสิ้นสุดที่ Lag 1 หรือ Cut off after lag 1 
  
**(2) แบบจำลอง Autoregressive ลำดับที่ 2**
แบบจำลอง AR(2) อยู่ในรูป
$$X_t=\alpha_0+\alpha_1X_{t-1}+\alpha_2X_{t-2}+\epsilon_t$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise </center>

- การที่ AR(2) มีเพียง Lag 2 เท่านั้นค่า SPAC จึงสิ้นสุดที่ Lag 2 หรือ Cut off after lag 2

**(3) แบบจำลอง Autoregressive ลำดับที่ p**
แบบจำลอง AR(p) อยู่ในรูป
$$X_t=\alpha_0+\alpha_1X_{t-1}+\alpha_2X_{t-2}+...+\alpha_pX_{t-p}+\epsilon_t$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise </center>

- การที่ AR(p) มีเพียง Lag p เท่านั้นค่า SPAC จึงสิ้นสุดที่ Lag p หรือ Cut off after lag p

### **3.2.2 แบบจำลอง Moving Average (MA)** 
**(1) แบบจำลอง Moving Average ลำดับที่ 1** \
แบบจำลอง MA(1) อยู่ในรูป
$$X_t=\beta_0+\epsilon_t-\beta_1\epsilon_{t-1}$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise </center>
    
จาก **Assumption : Stationary** จะได้ว่า
$$\mu=\beta_0$$
<center>$\gamma_0=(1+\beta_1^2)\sigma^2$ </center>
\begin{equation}
  \rho_k=\begin{cases}
    \frac{-\beta_1}{1+\beta_1^2}, & \text{if $k=1$}.\\
    0, & \text{otherwise}.
  \end{cases}
\end{equation}
    
- ค่า TAC ของ MA(1) จะมีค่าไม่เป็นศูนย์ที่ Lag 1 ส่วน Lag ที่เหลือจะมีค่า TAC เป็นศูนย์
    
- กล่าวว่า TAC สิ้นสุดหลังจาก 1 ช่วงเวลาที่แล้ว (Cuts off after Lag 1)

- สามารถพิสูจน์ได้ว่า $|\beta_1|<1$ ทำให้ MA(1) เทียบเท่า AR($\infty$) และได้ว่า TPAC มีค่าลดลงอย่าง Exponential

**(2) แบบจำลอง Moving Average ลำดับที่ 2** \
แบบจำลอง MA(2) อยู่ในรูป
$$X_t=\beta_0+\epsilon_t-\beta_1\epsilon_{t-1}-\beta_2\epsilon_{t-2}$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise </center>
    
จาก **Assumption : Stationary** จะได้ว่า
$$\mu=\beta_0$$
<center>$\gamma_0=(1+\beta_1^2+\beta_2^2)\sigma^2$ </center>
\begin{equation}
  \rho_k=\begin{cases}
    \frac{-\beta_1(1-\beta_2)}{1+\beta_1^2+\beta_2^2}, & \text{if $k=1$}.\\
    \frac{-\beta_2}{1+\beta_1^2+\beta_2^2}, & \text{if $k=2$}.\\
    0, & \text{k>2}.
  \end{cases}
\end{equation}

- กล่าวว่า TAC สิ้นสุดหลังจาก 2 ช่วงเวลาที่แล้ว (Cuts off after Lag 2)
- กล่าวว่า TPAC จะลดลงอย่าง Exponential ซึ่งสามารถพิสูจน์ได้จาก Root of Solution $\beta(L)=0$

**(3) แบบจำลอง Moving Average ลำดับที่ q**
แบบจำลอง MA(q) อยู่ในรูป
$$X_t=\beta_0+\epsilon_t-\beta_1\epsilon_{t-1}-\beta_2\epsilon_{t-2}-...-\beta_q\epsilon_{t-q}$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise </center>

- กล่าวว่า TAC สิ้นสุดหลังจาก q ช่วงเวลาที่แล้ว (Cuts off after Lag q)
- กล่าวว่า TPAC จะลดลงอย่าง Exponential

### **3.2.3 แบบจำลอง Autoregressive Moving Average (ARMA)** 
เพื่อลดปัญหาเรื่องจำนวนพารามิเตอร์ในแบบจำลอง AR และ MA ที่มีลำดับสูง  แบบจำลอง AR และ MA ถูกนำมารวมเข้าด้วยกันเป็นแบบจำลอง ARMA 

**(1) แบบจำลอง Autoregressive Moving Average ลำดับที่ (1,1)** \
แบบจำลอง ARMA(1,1) อยู่ในรูป
$$X_t=\alpha_0+\alpha_1X_{t-1}+\epsilon_t-\beta_1\epsilon_{t-1}$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise </center>
    
จาก **Assumption : Stationary** จะได้ว่า
$$\mu=\frac{\alpha_0}{1-\alpha_1}$$
<center>$\gamma_0=\frac{(1+\beta_1^2-2\alpha_1\beta_1)}{1-\alpha_1^2}\sigma^2$ </center>
\begin{equation}
  \rho_k=\begin{cases}
    \frac{(\alpha_1-\beta_1)(1-\alpha_1\beta_1)}{1+\beta_1^2-2\alpha_1\beta_1}, & \text{if $k=1$}.\\
    \alpha_1\rho_{k-1}, & \text{k $\ge$ 2}.
  \end{cases}
\end{equation}
    
- เงื่อนไข Stationary อยู่ในแนวคิดแบบเดียวกับ AR(1) และ MA(1) คือ รากของสมการ $1-\alpha_1L=0$ และ $1-\beta_1L=0$ ต้องมากกว่าหนึ่ง

- ค่า TAC ของ ARMA(1,1) มีลักษณะแบบ AR(1) คือลดลงแบบ Exponential
    
- ค่า TPAC ของ ARMA(1,1) มีลักษณะแบบ MA(1) คือลดลงแบบ Exponential

**(2) แบบจำลอง Autoregressive Moving Average ลำดับที่ (p,q)** \
แบบจำลอง ARMA(p,q) อยู่ในรูป
$$X_t=\alpha_0+\alpha_1X_{t-1}+\alpha_2X_{t-2}+...+\alpha_pX_{t-p}+\epsilon_t-\beta_1\epsilon_{t-1}-\beta_2\epsilon_{t-2}-...-\beta_q\epsilon_{t-q}$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise </center>
สามารถเขียนแบบจำลอง ARMA(p,q) อีกรูปแบบหนึ่งได้ดังนี้
$$\alpha(L)X_t=\alpha_0+\beta(L)\epsilon_t$$

โดยที่
$$ \alpha(L) = 1 - \alpha_1L - \alpha_2L^2 - ... - \alpha_pL^p \ $$
$$ \beta(L) = 1 - \beta_1L - \beta_2L^2 - ... - \beta_pL^p \ $$
    
จาก **Assumption : Stationary** ทำให้เกิดเงื่อนไขว่ารากของสมการ $\alpha(L)=0 $ และ $\beta(L)=0 $ ต้องมากกว่าหนึ่ง

- จึงทำให้ค่า TAC และ TPAC ของแบบจำลอง ARMA(p,q) มีการลดตัวแบบ Exponential

# **บทที่ 4. วิธีการของ Box-Jenkins: การประมาณค่าพารามิเตอร์และการตรวจสอบแบบจำลอง** <a class="anchor" id="4"></a>

[สารบัญ](#0.1)

## **4.1 การประมาณค่าพารามิเตอร์ (Parameter Estimation)**
### **4.1.1 การประมาณค่าพารามิเตอร์ในแบบจำลอง Autoregressive** 
แบบจำลอง AR(p) อยู่ในรูป
$$X_t=\alpha_0+\alpha_1X_{t-1}+\alpha_2X_{t-2}+...+\alpha_pX_{t-p}+\epsilon_t$$
<center>โดยที่ $\epsilon_t$ คือ Gaussian white noise, $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$ </center>
<center>และ $cov(\epsilon_{t_i},\epsilon_{t_i}) = 0 ; t_i \neq t_j $</center>
ในเงื่อนไขนี้ ทำให้ AR(p) เป็น Unbias and Consistency Estimator


### **4.1.2 การประมาณค่าพารามิเตอร์ในแบบจำลอง Moving Average** 
พิจารณาแบบจำลอง MA(1) ในรูปอย่างง่ายที่ $\beta_0= 0 $
$$ X_t=\epsilon_t-\beta_1\epsilon_{t-1} $$
แต่ $\epsilon_t$ คือ Gaussian white noise, $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$ ซึ่งไม่สามารถสังเกตุได้
การหาค่าพารามิเตอร์ $\beta_1$ การทำได้สองวิธีคือ

**1. Conditional Maximum Likelihood / Least Square** \
จากค่า
$$ln\mathcal{L}(\beta_1,\sigma^2;X_1,X_2,...,X_T) = -\frac{T}{2}ln(2\pi)-\frac{T}{2}ln(\sigma^2)-\frac{1}{2\sigma^2}\sum_{t=1}^{T}\epsilon_t^2$$
**Assume :** $\epsilon_0=0 $ สามารถพิสูจน์ได้ว่า
$$\epsilon_t=X_t+\beta_1X_{t-1}+\beta_1^2X_{t-2}+\beta_1^3X_{t-3}+...+\beta_1^{T-1}X1$$
                                                                                                                                           

**2. Unconditional Maximum Likelihood / Least Square** \
เขียน MA(1) ในรูปแบบย้อนกลับ
$$ X_1=b_t-\beta_1b_{t+1} $$ 
<center>หรือ $b_t=X_t+\beta_1b_{t+1}$</center>
การที่ {$X_t;t=1,..,T$} เป็นข้อมูลที่สังเกตุได้ และ $\beta_1$ เป็นค่าพารามิเตอร์ ดังนั้น
$$b_1,b_2,...,b_T \neq 0$$
ในขณะที่
$$b_0,b_{-1},..,b_{-\infty}\mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$$ 
จะได้ว่า
$$\hat{X_0}=\mathbb{E}[X_0|X] =\mathbb{E}[b_0+\beta_1b_1|X_1,...,X_T]=\beta_1b_1$$
และ
$$ \hat{\epsilon_0}=\mathcal{E}[\epsilon_0|X_1,...,X_T]=\mathcal{E}(X_0-\beta_1\epsilon_{-1}|X_1,...,X_T)=\mathcal{E}[X_0|X]=\hat{X_0}$$
จากนั้นใช้ค่า $\hat{\epsilon_0}$ หาค่าพารามิเตอร์ $\beta_1$ แทนที่ข้อสมมุติฐานใน Conditional Maximum Likelihood/ Least Square



### **4.1.3 การประมาณค่าพารามิเตอร์ในแบบจำลอง Autoregressive Moving Average** 
แบบจำลอง ARMA(p,q) อยู่ในรูป
$$X_t=\alpha_0+\alpha_1X_{t-1}+\alpha_2X_{t-2}+...+\alpha_pX_{t-p}+\epsilon_t-\beta_1\epsilon_{t-1}-\beta_2\epsilon_{t-2}-...-\beta_q\epsilon_{t-q}$$
โดยที่ อนุกรมเวลา $X_t;t=1,...,T$ เป็นข้อมูลที่สังเกตุได้ ส่วน $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$ ไม่สามารถสังเกตุได้ \
การหาค่าพารามิเตอร์จึงทำได้สองวิธีคือ Conditional และ Unconditional Maximum Likelihood / Least Square

## **4.2 การตรวจสอบแบบจำลอง (Diagnostic Checking)**
ตามแนวคิดของ Box-Jenkins แล้ว
1. ค่า Residual, $\epsilon_t\mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2) $ และ
2. $ cov(\epsilon_{t_i},\epsilon_{t_j})=0$ ซึงตรวจสอบได้จาก Ljung-BoxQ หรือแผนภูมิ Autocorrelation
3. ค่าสัมประสิทธิของแบบจำลองต้องมีนัยสำคัญเชิงสถิติทุกตัว

มิฉะนั้นต้องกลับไป Train หรือ Fit แบบจำลองใหม่จนกว่าเงื่อนไขทั้งสามข้อจะเป็นจริง

ในกรณีมีหลายแบบจำลองที่ถูกต้องตามแนวคิดของ Box-Jenkins แล้วสามารถเลือกแบบจำลองที่มีค่า RMSE น้อยที่สุด หรือจะเลือกแบบจำลองตามแนวคิดของ AIC และ BIC ดังนี้

**1. Akaike's Information Criterion (AIC)**
   $$ AIC(k) = -2(\frac{l}{N})+\frac{2}{N}k $$
โดยที่ $l$ คือ Log Likelihood, N คือจำนวน Sample และ k คือจำนวนพารามิเตอร์

- แบบจำลองที่ดีความีค่า Log-Likelihood มาก นั่นคือค่า AIC(k) ยิ่งต่ำยิ่งได้แบบจำลองที่ดี
- ค่า AIC มีข้อเสียคือจำนวน Sample ยิ่งมากแล้วค่า AIC ยิ่งน้อย ดังนั้นในกรณีที่ N มีค่ามากจะเลือกใช้แนวทางของ BIC ดังต่อไปนี้

**2. Specifies the Bayesian Information Criterion (BIC) หรือ Schwarz's Bayesian Criterion (SBC)**
   $$ BIC(k) = -2(\frac{l}{N})+\frac{lnN}{N}k $$
- ค่า BIC จะลงโทษแบบจำลองที่มีจำนวนพารามิเตอร์มาก เพราะยิ่งพารามิเตอร์มากขึ้นค่า BIC จะเพิ่มขึ้น


# **บทที่ 5. แบบจำลองอนุกรมเวลาที่ไม่มีความนิ่ง (Nonstationary Time Series Models)** <a class="anchor" id="4"></a>

[สารบัญ](#0.1)

## **5.1 อนุกรมเวลาที่ไม่มีความนิ่ง (Nonstationary Time Seires Models)**
อนุกรมเวลาจะเป็น Stationary ก็ต่อเมื่อ \

1. $\mathbb{E}[X_t]=\mu \sim const; t=1,2,...T$
2. $var(X_t)=\sigma_x^2\sim;t=1,2,...,T$
3. $cov(X_{t_i},X_{t_j})=\gamma_{\tau}\sim const;\tau=$distance from $t_i$ to $t_j$

หาก $X_t$ ไม่เป็น stationary แล้ว $X_t$ อาจมีรูปแบบต่อไปนี้
$$X_t=\mu_t+\epsilon_t$$
<center>โดยที่ $\mu_t$ คือ Trend ของอนุกรมเวลา และ $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$ </center>

### **5.1.1 อนุกรมเวลา $X_t$ ประกอบด้วยแนวโน้มแบบกำหนดได้ (Deterministic Trend)**
ในกรณีนี้ $\mu_t$ คือ Deterministic Trend ซึ่งเป็นค่าเฉลี่ยของ $X_t$ กรณีที่ $\mu_t$ อยู่ในรูปแบบเชิงเส้นจะได้ว่า
$$\mu_t=\varphi_0+\varphi_1t $$
นั่นคือ $X_t$ จะอยู่ในรูป
$$X_t=\varphi_0+\varphi_1t+\epsilon_t$$
จึงทำให้ $X_t$ เป็น Nonstationary ดังนี้
$$\mathbb{E}[X_t]=\varphi_0+\varphi_1t $$
$$var(X_t)=\sigma_\epsilon^2$$
$$cov(X_{t_i},X_{t_j})=cov(\epsilon_{t_i},\epsilon_{t_j})=0$$
ใช้ First Difference หรือ $\Delta X_t:=X_t-X_{t-1}$ เพื่อกำจัด Trend ให้ออกจากอนุกรมเวลา
$$\Delta X_t = \varphi_1+(\epsilon_t-\epsilon_{t-1}) $$
$\Delta X_t$ มีความหมายว่า การเบี่ยงเบนออกจากเส้นแนวโน้มที่เวลา $t-1$ จะเบียงเบนกลับสู่เส้นแนวโน้มเดิมที่เวลา $t$ ด้วยขนาดเท่ากับ $\epsilon_{t-1}$ จากนั้น ณ เวลา $t$ เดียวกันนี้ $X_t$ จะพยายามเบี่ยงเบนออกจากเส้นแนวโน้มด้วยขนาด $\epsilon_{t}$  นอกจากนี้ยังพบว่า $\Delta X_t$ มีลักษณะของ Stationary ดังนี้
$$\mathbb{E}[\Delta X_t]=\varphi_1 \sim const $$
$$var(\Delta X_t)=2\sigma_\epsilon^2 \sim const$$
$$cov(\Delta X_{t},\Delta X_{t-1})=cov(\epsilon_t-\epsilon_{t-1},\epsilon_{t-1}-\epsilon_{t-2})=\sigma_\epsilon^2 \sim const$$ 

การกำจัด Trend ออกจากอนุกรมเวลา $X_t$ แล้วได้อนุกรมเวลาที่มีความนิ่ง  อนุกรมเวลา $X_t$ ประเภทนี้เรียนกว่า **Trend Stationary**

### **5.1.2 อนุกรมเวลา $X_t$ ประกอบด้วยแนวโน้มแบบสุ่ม (Stochastic Trend)**
ในกรณีนี้ $\mu_t$ จะเป็นค่าสุ่ม (random) โดยมีลักษณะดังนี้
$$\mu_t-\mu_{t-1}= \varphi_0+\nu_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(\varphi_0,\sigma_\nu^2)$$
ค่า $\varphi_0$ คือค่าแนวโน้มของการสุ่มซึ่งแบ่งออกเป็นสองกรณีดังนี้

**(1) แนวโน้มที่เป็นแบบสุ่มไม่มีค่าคงที่** 

ในกรณีนี้ $\varphi_0=0 $ ดังนั้น
$$\mu_t=\mu_{t-1}+\nu_t $$
ซึ่งเขียนในรูปทั่วไปได้ดังนี้
$$\mu_t=\sum_{i=1}^t\nu_i\mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,t\sigma_\nu^2)$$
จึงทำให้ $X_t$ เป็น Nonstationary ดังนี้
$$\mathbb{E}[X_t]=0 $$
$$var(X_t)=t\sigma_\nu^2+\sigma_\epsilon^2$$
$$cov(X_{t},X_{t-1})=cov(\sum_{i=1}^t\nu_i+\epsilon_{t},\sum_{i=1}^{t-1}\nu_i+\epsilon_{t-1})=var(\sum_{i=1}^{t-1}\nu_i)=(t-1)\sigma_\nu^2$$
ใช้ First Difference หรือ $\Delta X_t:=X_t-X_{t-1}$ เพื่อกำจัด Trend ให้ออกจากอนุกรมเวลาจะได้ว่า
$$ \Delta X_t=\nu_t+(\epsilon_t-\epsilon_{t-1})$$
นอกจากนี้ยังพบว่า $\Delta X_t$ มีลักษณะของ Stationary ดังนี้
$$\mathbb{E}[\Delta X_t]=0 \sim const $$
$$var(\Delta X_t)=\sigma_\nu^2+2\sigma_\epsilon^2 \sim const$$
$$cov(\Delta X_{t},\Delta X_{t-1})=cov(\nu_t+(\epsilon_t-\epsilon_{t-1}),\nu_{t-1}+(\epsilon_{t-1}-\epsilon_{t-2}))=var(\epsilon_{t-1})=\sigma_\epsilon^2 \sim const$$ 

**(2) แนวโน้มที่เป็นแบบสุ่มมีค่าคงที่** 

ในกรณีนี้ $\varphi_0\neq0 $ ดังนั้น
$$\mu_t=\mu_{t-1}+\varphi_0+\nu_t $$
เขียนในรูปทั่วไปได้คือ
$$\mu_t=\varphi_0t+\sum_{i=1}^t\nu_i\mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(\varphi_0t,t\sigma_\nu^2)$$
จึงทำให้ $X_t$ เป็น Nonstationary ดังนี้
$$\mathbb{E}[X_t]=\varphi_0t $$
$$var(X_t)=t\sigma_\nu^2+\sigma_\epsilon^2$$
$$cov(X_{t},X_{t-1})=cov(\sum_{i=1}^t\nu_i+\epsilon_{t},\sum_{i=1}^{t-1}\nu_i+\epsilon_{t-1})=var(\sum_{i=1}^{t-1}\nu_i)=(t-1)\sigma_\nu^2$$
ใช้ First Difference หรือ $\Delta X_t:=X_t-X_{t-1}$ เพื่อกำจัด Trend ให้ออกจากอนุกรมเวลาจะได้ว่า
$$ \Delta X_t=\varphi_0+\nu_t+(\epsilon_t-\epsilon_{t-1})$$
นอกจากนี้ยังพบว่า $\Delta X_t$ มีลักษณะของ Stationary ดังนี้
$$\mathbb{E}[\Delta X_t]=\varphi_0 \sim const $$
$$var(\Delta X_t)=\sigma_\nu^2+2\sigma_\epsilon^2 \sim const$$
$$cov(\Delta X_{t},\Delta X_{t-1})=cov(\nu_t+(\epsilon_t-\epsilon_{t-1}),\nu_{t-1}+(\epsilon_{t-1}-\epsilon_{t-2}))=var(\epsilon_{t-1})=\sigma_\epsilon^2 \sim const$$ 

### **5.1.3 อนุกรมเวลา $X_t$ เป็นการเดินแบบสุ่ม (Random Walk)**

อนุกรมเวลา $X_t$ มีการเคลื่อนที่แบบ Random Walk หมายความว่า
$$ X_t=X_{t=1}+\epsilon_t $$
<center>โดยที่ $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$</center>
เมื่อกำหนดให้ $\epsilon_0=0$ แล้วอนุกรมเวลา $X_t$ สามารถเขียนใหม่ได้ดังนี้

$$X_t=\sum_{i=1}^t\epsilon_i $$

จึงทำให้ $X_t$ เป็น Nonstationary ดังนี้
$$\mathbb{E}[X_t]=0 $$
$$var(X_t)=t\sigma_\epsilon^2$$
$$cov(X_{t},X_{t-1})=cov(\sum_{i=1}^t\epsilon_i,\sum_{i=1}^{t-1}\epsilon_i)=var(\sum_{i=1}^{t-1}\epsilon_i)=(t-1)\sigma_\epsilon^2$$

ใช้ First Difference หรือ $\Delta X_t:=X_t-X_{t-1}$ เพื่อกำจัด Trend ให้ออกจากอนุกรมเวลาจะได้ว่า
$$ \Delta X_t=\epsilon_t$$

นอกจากนี้ยังพบว่า $\Delta X_t$ มีลักษณะของ Stationary ดังนี้
$$\mathbb{E}[\Delta X_t]=0 \sim const $$
$$var(\Delta X_t)=\sigma_\epsilon^2 \sim const$$
$$cov(\Delta X_{t},\Delta X_{t-1})=cov(\epsilon_t,\epsilon_{t-1})=0 \sim const$$ 


หากกำหนดให้ $X_0=0$ และ $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,1)$เป็นทางการเงินการเคลื่อนที่ของอนุกรมเวลาแบบ Random Walk นี้คือ Arithmatic Brownian Motion
$$\mathbb{E}[X_t]=0 \sim const $$
$$var(X_t)=t$$
$$cov(X_{t},X_{t-1})=cov(\sum_{i=1}^t\epsilon_i,\sum_{i=1}^{t-1}\epsilon_i)=var(\sum_{i=1}^{t-1}\epsilon_i)=(t-1)$$

ในกรณีที่ $X_t$ เป็นการเคลื่อนที่แบบสุ่มที่มีแนวโน้ม หรือ Random Walk with Drift

$$ X_t=X_{t=1}+\varphi_0+\epsilon_t $$
<center>โดยที่ $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$</center>

จะได้ว่า

$$X_t=\varphi_0t+\sum_{i=1}^t\epsilon_i $$

จึงทำให้ $X_t$ เป็น Nonstationary ดังนี้
$$\mathbb{E}[X_t]=\varphi_0t $$
$$var(X_t)=t\sigma_\epsilon^2$$
$$cov(X_{t},X_{t-1})=cov(\sum_{i=1}^t\epsilon_i,\sum_{i=1}^{t-1}\epsilon_i)=var(\sum_{i=1}^{t-1}\epsilon_i)=(t-1)\sigma_\epsilon^2$$

ใช้ First Difference หรือ $\Delta X_t:=X_t-X_{t-1}$ เพื่อกำจัด Trend ให้ออกจากอนุกรมเวลาจะได้ว่า
$$ \Delta X_t=\varphi_0+\epsilon_t$$

นอกจากนี้ยังพบว่า $\Delta X_t$ มีลักษณะของ Stationary ดังนี้
$$\mathbb{E}[\Delta X_t]=\varphi_0 \sim const $$
$$var(\Delta X_t)=\sigma_\epsilon^2 \sim const$$
$$cov(\Delta X_{t},\Delta X_{t-1})=cov(\epsilon_t,\epsilon_{t-1})=0 \sim const$$ 

หากกำหนดให้ $X_0=0$ และ $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,1)$เป็นทางการเงินการเคลื่อนที่ของอนุกรมเวลาแบบ Random Walk นี้คือ Arithmatic Brownian Motion with drift
$$\mathbb{E}[X_t]=\varphi_0t \sim const $$
$$var(X_t)=t$$
$$cov(X_{t},X_{t-1})=cov(\sum_{i=1}^t\epsilon_i,\sum_{i=1}^{t-1}\epsilon_i)=var(\sum_{i=1}^{t-1}\epsilon_i)=(t-1)$$

### **5.1.4 อนุกรมเวลา $X_t$ อยู่ในรูปผลรวมลำดับที่ d (Integrated of order d)**
หัวข้อ 5.1.1-5.1.3 ได้ว่า First Difference สามาารถทำให้ $X_t$ กลับมาเป็น Stationary ดังนี้


**กรณี Deterministic Trend:** 
<center>$\Delta X_t = \varphi_1+(\epsilon_t-\epsilon_{t-1})= \varphi_1+a_t$ where $a_t:= \epsilon_t-\epsilon_{t-1}$</center>

**กรณี Stochastic Trend:** 
<center>$\Delta X_t=\varphi_0+\nu_t+(\epsilon_t-\epsilon_{t-1})= \varphi_0+a_t$ where $a_t:= \nu_t+(\epsilon_t-\epsilon_{t-1})$</center>

**กรณี Random Walk:** 
<center>$ \Delta X_t=\varphi_0+\epsilon_t$</center>

หมายความว่า First Differenc ของ $X_t$ หรือ $\Delta X_t$ ทำให้อนุกรมเวลา X_t เป็น Stationary เขียนแทนว่า

$$\Delta X_t=\alpha_0+\epsilon_t \rightarrow  X_t \sim I(1)$$
ซึ่งก็คือ $X_t$ อยู่ในรูปผลรวมลำดับหนึ่ง หรือ $I(1)$

ทำนองเดียวกันหาก $X_t$ เป็น Nonstationary แต่ Second Difference หรือ $\Delta^2 X_t$ ทำให้อนุกรมเวลา $X_t$ เกิดความนิ่ง

นั่นคือ <center>$\Delta^2 X_t=\alpha_0+\epsilon_t \rightarrow  X_t \sim I(2)$</center>
ซึ่งก็คือ $X_t$ อยู่ในรูปผลรวมลำดับสอง หรือ $I(2)$

สามารถกล่าวในรูปทั่วไปได้ว่า d-order Difference หรือ $\Delta^dX_t$ สามารถทำให้อนุกรมเวลา $X_t$ เป็น Stationary เขียนแทนว่า
$$\Delta^d X_t=\alpha_0+\epsilon_t \rightarrow  X_t \sim I(d)$$
ซึ่งก็คือ $X_t$ อยู่ในรูปผลรวมลำดับ d หรือ $I(d)$ ซึ่งเป็นเพราะ $X_t$ เกิดจากผลรวมจำนวน d ครั้งซ้อนของอนุกรมเวลาที่มีความนิ่ง

## **5.2 ค่า TAC และ TPAC ของอนุกรมเวลาที่มีการเดินแบบสุ่ม**
ในกรณี AR(1): $X_t = \alpha_0+\alpha_1X_1-t+\epsilon_t$ จะได้ว่า
$$\rho_k=\alpha_1^k$$ 
\begin{equation}
  \phi_kk=\begin{cases}
    \rho_1, & \text{if $k=1$}.\\
    0, & \text{if $k \ge 2$}.
  \end{cases}
\end{equation}

ทำนองเดียวกันกรณี Random Walk: $X_t=X_{t-1}+\epsilon_t$ จะได้ว่า
$$\rho_k=1$$ 
\begin{equation}
  \phi_kk=\begin{cases}
    1, & \text{if $k=1$}.\\
    0, & \text{if $k \ge 2$}.
  \end{cases}
\end{equation}

หมายความว่า TAC จะมีค่าเป็น 1 คงที่เสมอ ส่วน TPAC จะมีค่าเป็น 1 สำหรับช่วงเวลาแรก จากนั้น TPAC จะมีค่าเป็นศูนย์ในทันที

## **5.3 แบบจำลอง Autoregressive Integrated Moving Average (ARIMA)**
ในแบบจำลอง ARMA(p,q) นั้นอนุกรมเวลา $X_t$ ต้องมีคุณสมบัติ Stationary มิฉะนั้นต้องแปลง $X_t$ ให้มีคุณสมบัติ Stationary เช่นการใช้ First Differncing ก่อนจึงค่อยสร้างแบบจำลอง ARMA

หากกำหนดให้ $X_t\sim I(d)$ ดังนั้น $\Delta^d X_t$ จะมีความนิ่ง และสามารถใช้แบบจำลอง ARMA(p,q) ได้ และเรียกแบบจำลองประเภทนี้ว่า **Autoregressive Integrated Moving Average** หรือ **ARIMA(p,d,q)**

แบบจำลอง ARIMA(p,d,q) แสดงได้ดังนี้
$$ \alpha(L)\Delta^d X_t=\theta_0+\beta(L)\epsilon_t$$
โดยที่ $\theta_0$ คือสัมประสิทธิ์ของของเส้นแนวโน้มซึ่งมีค่าคงที่
$$\alpha(L)=1-\alpha_1L-\alpha_2L^2-...-\alpha_pL^p$$
$$\beta(L)=1-\beta_1L-\beta_2L^2-...-\beta_qL^q$$

**ตัวอย่าง :** แบบจำลอง ARIMA(0,1,1)

ได้ว่า $\alpha(0)=1$ และ $\beta(1)=1-\beta_1L$ ดังนั้น ARIMA(0,1,1) อยู่ในรูป
$$1\Delta X_t=\theta_0+(1-\beta_1L)\epsilon_t$$
$$X_t-X_{t-1}=\theta_0+(1-\beta_1L)\epsilon_t$$
จะได้ว่าอนุกรมเวลา $X_t$ คือ Random Walk with Drift
$$X_t=X_{t-1}+\theta_0+\epsilon_t-\beta_1\epsilon_{t-1}$$

## **5.4 การทดสอบความนิ่งของอนุกรมเวลา (Stationary Test of Time Series)**

การทดสอบแบบแรกชื่อ Dickey-Fuller หรือ DF-test ใช้ทดสอบเฉพาะแบบจำลอง AR(1)
จากแบบจำลอง AR(1)
$$X_t=\rho X_{t-1}+\epsilon_t$$
อนุกรมเวลาจะมีความนิ่งก็ต่อเมื่อ $|\rho|<1$ และจะเป็น Random Walk เมื่อ $\rho=1$

ดังนั้นสมมุติฐานที่ใช้ในการทดสอบคือ 
$$H_0:\rho=1$$
$$H_1:|\rho|<1$$
และให้ชื่อการทดสอบ $\rho=1$ นี้ว่าการทดสอบ **Unit Root Test**

การทดสอบ DF-test ได้รับการพัฒนาให้ตรวจสอบ Random Walk ได้ถึงสามรูปแบบคือ

1. $X_t=\rho X_{t-1}+\epsilon_t$
2. $X_t=\beta_0+\rho X_{t-1}+\epsilon_t$
3. $X_t=\beta_0+\beta_1 t+\rho X_{t-1}+\epsilon_t$

นำ $X_{t-1}$ ลบออกทั้งสองข้างของสมการจะได้ว่า


1. $\Delta X_t=(\rho-1) X_{t-1}+\epsilon_t=\gamma X_{t-1}+\epsilon_t$
2. $\Delta X_t=\beta_0+(\rho-1) X_{t-1}+\epsilon_t=\beta_0+\gamma X_{t-1}+\epsilon_t$
3. $\Delta X_t=\beta_0+\beta_1 t+(\rho-1)\rho X_{t-1}+\epsilon_t=\beta_0+\beta_1 t+\gamma X_{t-1}+\epsilon_t$

โดยที่่ $\gamma=\rho-1$

สมมุติฐานของ **Unit Root Test** จึงเปลี่ยนมาเป็น

$$H_0:\gamma=0$$
$$H_1:\gamma<0$$

โดย student t-stat $t^*=\frac{\hat{\gamma}}{se(\hat{\gamma})} $


## **5.5 การทดสอบความนิ่งของอนุกรมเวลาด้วยวิธี Augmented Dickey-Fuller (ADF-Test)**

DF-Test ใช้ได้เฉพาะ AR(1) เท่านั้นแต่การทดสอบ ADF-Test สามารถทดสอบได้ทุกๆ AR(p) โดยอนุกรมแบบ Random Walk สามารถมีได้ถึงสามรูปแบบดังนี้

1. $\Delta X_t=\gamma X_{t-1}+\sum_{i=1}^{p-1}c_i \Delta X_{t-i}+\epsilon_t$
2. $\Delta X_t=\beta_0+\gamma X_{t-1}+\sum_{i=1}^{p-1}c_i \Delta X_{t-i}+\epsilon_t$
3. $\Delta X_t=\beta_0+\beta_1 t + \gamma X_{t-1}+\sum_{i=1}^{p-1}c_i \Delta X_{t-i}+\epsilon_t$

สมมุติฐานหลักของ ADF-Test คือแบบจำลองเป็น **Unit Root** ดังนี้
$$H_0:\gamma=0$$
$$H_1:\gamma<0$$

- เมื่อมีความผันแปรทางฤดูกาลในอนุกรมเวลาจะทำให้แบบจำลองที่ใข้มีความซับซ้อนขึ้นเช่น อนุกรมเวลาในช่วงเวลา $t$ กับช่วงเวลา $t-2$ อาจมึความสัมพันธ์กัน



## **6.1 การกำจัดความผันแปรทางฤดูกาลออกไปจากอนุกรมเวลา**

ความผันแปรทางฤดูกาลเป็นแบบกำหนดได้ (Deterministic Seasonal)

- ความผันแปรทางฤดูกาลมีรูปแบบที่แน่นอนในแต่ละปี และ เป็นอิสระจากส่วนประกอบอื่นๆ ในอนุกรมเวลา


### **6.1.1 การใช้ตัวแปรหุ่น (Dummy Variables)**
ในกรณีที่ Seasonal Variation เป็น Deterministic จะเขียนอนุกรมเวลาได้ดังนี้

$$X_t=\beta_0+S_t+\epsilon_t$$
โดยที่
- $S_t$ คือส่วนของความผันแปรทางฤดูกาล 
- $\epsilon_t$ คือ Error term ที่ประกอบด้วย Trend, Cycle และ Irregular

แสดงส่วนของ Seasonal ด้วยผลรวมเชิงเส้น (Linear Combination) ของตัวแปรหุ่น (Dummy Variable, $D_{jt}$)
$$S_t=\sum_{j=1}^{s-1}\beta_jD_{jt}$$
โดยที่
\begin{equation}
  D_{jt}=\begin{cases}
    1, & \text{เมื่อ t คือฤดูกาลที่ j}.\\
    0, & \text{กรณีอื่นๆ}.
  \end{cases}
\end{equation}
ทำให้ $X_t$ อยู่ในรูปแบบสมการถดถอยดังนี้

$$X_t=\beta_0+(\sum_{j=1}^{s-1}\beta_jD_{jt})+\epsilon_t$$


### **6.1.2 การใช้ผลต่างทางฤดูกาล (Seasonal Differenceing)**
การที่รูปแบบทางฤดูกาลมีค่าคงที่ในแต่ละปี ทำให้การ Differencing สามารถกำจัดอิทธิพลของฤดูกาลออกจากอนุกรมเวลาได้

ผลต่างลำดับที่ $s$ ของ $X_t$ สามารถแสดงได้ดังนี้

$$ \Delta_sX_t=X_t-X_{t-s}=(1-L^s)X_t$$

ผลต่างฤดูกาล ต้องให้ข้อมูลกำกับเสมอด้วยว่าช่วงเวลาของฤดูกาล ($s$) มีค่าเท่าไรเสมอ
ตัวอย่างเช่น : อนุกรมเวลาเป็นข้อมูลรายไตรมาส ($s$=4) การกำจัดความผันแปรทางฤดูกาลทำได้โดย
$$\Delta_4X_t=X_t-X_{t-4}$$


## **6.2 แบบจำลองของ Box-Jenkins กับอิทธิพลของความผันแปรทางฤดูกาล**
หาก Seasonal Variation มีลักษณะสุ่ม (Stochastic Seasonal) และยังมีความสัมพันธ์อื่นๆ กับ $X_t$ แล้ว ทั้ง Dummry Variable และ Seasonal Differencing ไม่สามารถขจัดความผันแปรทางฤดูกาลได้ แต่สามารถใส่ฤดูกาลในแบบจำลองของ Box-Jenkins โดยตรง

### **6.2.1 แบบจำลองกรณีความผันแปรทางฤดูกาลมีความนิ่ง**
อิทธิพลของฤดูกาลอยู่ในรูปแบบไม่แน่นอน  ส่งผลกระทบลดลงเมื่อเวลาผ่านไป

ตัวอย่างเช่น: มีการประท้วงในไตรมาสสองของปีที่แล้ว แต่ไม่เกิดในปีนี้ ดังนั้นการประท้วงในปีที่แล้วส่งผลต่อไตรมาสสองของปีนี้เพียงชั่วคราวเท่านั้น  นั่นคือความผันแปรทางฤดูกาลมีความนิ่ง (Stationary Seasonal Process)

กรณีเช่นนี้ไม่จำเป็นต้องขจัดผลของฤดูกาลออกจากแบบจำลอง แต่นำฤดูกาลเข้าไปร่วมในแบบจำลอง เรียกกว่า Seasonal Autoregressive Moving Average (SARMA)

แบบจำลอง SARMA(P,Q) หรือ ARMA(P,Q)$_s$ มีรูปแบบทั่วไปคือ

$$A(L^s)X_t=B(L^s)\nu_t$$

โดย $s$ คือช่วงเวลาของฤดูกาล และ

$$A(L^s)=1-A_1L^s-A_2L^{2S}-...-A_PL^{Ps}$$
$$B(L^s)=1-B_1L^s-B_2L^{2S}-...-B_QL^{Qs}$$
<center>$\nu_t$ คือ error term ที่มีทั้ง Trend, cycle และ Irregular</center>


**ตัวอย่าง** อนุกรมเวลา $X_t$ เป็นรายไตรมาส มีความผันแปรทางฤดูที่มีความนิ่ง สามารถเขียนได้ดังนี้
$$X_t=A_1X_{}t-4+\nu_t$$
สามารถพิสูจน์ได้ว่า ค่าเฉลี่ย ความแปรปรวน  TAC และ TPAC แสดงได้ดังนี้
$$\mathbb{E}[X_t]=\mu=0$$
$$var(X_t)=cov(X_t,X_t)=\gamma_0=\frac{\sigma^2}{1-A_1^2}$$
\begin{equation}
  \rho_k=\begin{cases}
    (A_1)^{k/4}, & \text{k=0,4,8,...}.\\
    0, & \text{กรณีอื่นๆ}.
  \end{cases}
\end{equation}
\begin{equation}
  \varphi_{k,k}=\begin{cases}
    \rho_4, & \text{k=4}.\\
    0, & \text{กรณีอื่นๆ}.
  \end{cases}
\end{equation}

- ลักษณะของค่า TAC จะลดลงอย่าง exponential ณ t=0,4,8,...
- ส่วนค่า TPAC จะมีค่าไม่เท่ากับศูนย์ ณ ช่วงเวลา t=4 เท่านั้น

ในทางปฎิบัติ $\nu_t$ อาจอยู่ในรูปแบบ ARMA(p,q) นั่นคือ
$$\alpha(L)\nu_t=\beta(L)\epsilon_t$$
จัดรูปใหม่จะได้ว่า<center>$\nu_t=\frac{\beta(L)}{\alpha(L)}\epsilon_t$</center>
<center>โดยที่ $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$</center>
โดย $\alpha(L)$ และ $\beta(L)$ คือ

$$\alpha(L)=1-\alpha_1L-\alpha_2L^{2}-...-\alpha_pL^{p}$$
$$\beta(L)=1-\beta_1L-\beta_2L^{2}-...-\beta_qL^{q}$$

แทนค่า $\nu_t$ ลงในแบบจำลอง ARMA(P,Q)$_s$ จะได้แบบจำลอง SARMA ในรูปทั่วไปหรือ ARMA(p,q)(P,Q)$_s$ คือ

$$A(L^s)\alpha(L)X_t=B(L^s)\beta(L)\epsilon_t$$

และให้ชื่อแบบจำลองนี้ว่า **Multiplicative Seasonal ARMA model**

**ตัวอย่าง** แบบจำลอง ARMA(0,1)(0,1)$_s$ สามารถเขียนได้ดังนี้
$$(1)(1)X_t=(1-B_1L)(1-\beta_1L)\epsilon_t$$
$$X_t=(1-\beta_1L-B_1L^s+\beta_1B_1L^{s+1})\epsilon_t $$
$$X_t=\epsilon_t-\beta_1\epsilon_{t-1}-B_1\epsilon_{t-s}+\beta_1B_1\epsilon_{t-s-1}$$
และสามารถพิสูจน์ได้ว่า
$$\mathbb{E}[X_t]=\mu=0$$
$$var(X_t)=cov(X_t,X_t)=\gamma_0=(1+\beta_1^2)(1+B_1^2)\sigma^2$$
\begin{equation}
  \rho_k=\begin{cases}
    \frac{-\beta_1}{1+\beta_1^2}, & \text{k=1}.\\
    \frac{-B_1}{1+B_1^2}, & \text{k=s}.\\
    \frac{\beta_1B_1}{(1+\beta_1^2)(1+B_1^2)}, & \text{k=s-1 และ s+1}.\\
    0, & \text{k $\neq$ 0,1,s-1,s,s+1}.
  \end{cases}
\end{equation}
จุดที่น่าสังเกตคือในกรณี Seasonal Variation เกิดที่ช่วงเวลา t=s แล้ว TAC จะไม่เป็น 0 ที่ช่วงเวลา 1,s-1,s,s+1 

### **6.2.2 แบบจำลองกรณีความผันแปรทางฤดูกาลไม่มีความนิ่ง**
อิทธิพลของฤดูกาลอยู่ในรูปแบบไม่แน่นอน  ผลของฤดูกาลไม่หายไปหรือหายไปอย่างช้าๆ

ตัวอย่างเช่น มีการประท้วงในไตรมาสที่ 2 ทำให้มียอดขายลดลง และในปีต่อไปยังพบว่าการประท้วงเกิดในไตรมาส 2 ซ้ำอีก และเหตุการณ์ประท้วงนี้ดูเหมือนจะไม่หายไปง่ายๆ นั่นคือความผันผวนทางฤดูกาบมีลักษณะไม่นิ่ง

จำเป็นต้องใช้ Seasonal Differencing ทำให้อิทธิพลของฤดูกาลมีความนิ่งเสียก่อน ดังนี้

$$A(L^s)\alpha(L)\Delta_s^DX_t=B(L^s)\beta(L)\epsilon_t$$
โดยที่ $D$ คือ ดีกรีในการทำผลต่างทางฤดูกาล

- $\Delta_s^DX_t = (1-L^s)^DX_t$ 
    - ตัวอย่างเช่น $\Delta_s^2X_t=(1-L^s)^2X_t=(1-2L^s+2L^{2s})X_t=X_t-2X_{t-s}+X_{t-2s}$
- $A(L^s)=1-A_1L^s-A_2L^{2s}-...-A_PL^{Ps}$
- $B(L^s)=1-B_1L^s-B_2L^{2s}-...-B_QL^{Qs}$
- $\alpha(L)=1-\alpha_1L-\alpha_2L^{2}-...-\alpha_pL^{p}$
- $\beta(L)=1-\beta_1L-B_2L^{2}-...-\beta_qL^{q}$

**ตัวอย่างเช่น** (p,q)(P,Q)$_s$ คือ (0,0)(0,2)$_4$ โดยที่ D=2 จะได้ว่าแบบจำลองคือ
$$A(L^0)\alpha(L^0)\Delta_4^2X_t=B(L^4)\beta(L^0)\epsilon_t$$
$$(1)(1)\Delta_4^2X_t=(1-B_1L^{1(4)}-B_2L^{2(4)})(1)\epsilon_t$$
$$\Delta_4^2X_t=(1-B_1L^{4}-B_2L^{8})(1)\epsilon_t$$
$$(1-L^4)^2X_t=(1-B_1L^{4}-B_2L^{8})(1)\epsilon_t$$
$$X_t-2X_{t-4}+X_{t-8}=\epsilon_t-B_1\epsilon_{t-4}-B_2\epsilon_{t-8}$$

หลังจากกำจัดผลของฤดูกาลออกแล้วอนุกรมเวลาย้งคงไม่นิ่ง จำเป็นต้องใช้ d-Differencing เพื่อกำจัดความไม่นิ่งที่ยังเหลืออยู่ และให้ชื่อแบบจำลองว่า **Seasonal Autoregressive Integrated Moving Average** ลำดับที่ (p,d,q)(P,D,Q)_$s$ หรือ $SARIMA(p,d,q)(P,D,Q)_s$

$$A(L^s)\alpha(L)\Delta^d\Delta_s^DX_t=B(L^s)\beta(L)\epsilon_t$$

**ตัวอย่าง** SARIMA(0,1,0)(0,1,1)$_{12}$
$$A(L^{12})\alpha(L)\Delta^1\Delta_{12}^1X_t=B(L^{12})\beta(L)\epsilon_t$$
$$(1)(1)(1-L)(1-L^{12})X_t=(1-B_1L^{1(12)})(1)\epsilon_t$$
$$(1-L)(1-L^{12})X_1=(1-B_1L^{12})\epsilon_t$$

**ตัวอย่าง** SARIMA(0,2,0)(0,2,2)$_{12}$
$$A(L^{12})\alpha(L)\Delta^2\Delta_{12}^2X_t=B(L^{12})\beta(L)\epsilon_t$$
$$(1)(1)(1-L)^2(1-L^{12})^2X_t=(1-B_1L^{1(12)}-B_2L^{2(12)})(1)\epsilon_t$$
$$(1-L)^2(1-L^{12})^2X_t=(1-B_1L^{12}-B_2L^{24)})\epsilon_t$$

**ตัวอย่าง** พิจารณาข้อมูลกำไรต่อเดือนของบริษัทหนึ่ง (ล้านบาท) อยู่ในไฟล์ชื่อ ch3_4.xlsx

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import matplotlib as mpl
import matplotlib.pyplot as plt   # data visualization


import pandas as pd 
Profit = pd.read_excel(r'input/ch3_4.xlsx')
Profit['Date'] = pd.to_datetime(Profit['Month']+Profit['Year'].astype(str))
Profit.head()
```

```python
mpl.rcParams['font.family'] = 'Tahoma'
plt.style.use('seaborn-v0_8-poster')
plt.plot(Profit['Date'],Profit['Profit'])
plt.title('ผลกำไรต่อเดือนของบริษัทแห่งหนึ่ง (หมื่นบาท)')
plt.xlabel('เดือน')
plt.ylabel('กำไร')
plt.axis('tight')
plt.grid('minor')
plt.grid('on')
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig, axes = plt.subplots(2,1, dpi= 100)
plot_acf(Profit['Profit'], lags=50, ax=axes[0],title="ค่า SAC ของผลกำไร $Profit_t$")
plot_pacf(Profit['Profit'], lags=50, ax=axes[1],title="ค่า PAC ของผลกำไร, $Profit_t$")
plt.axis('tight')
plt.grid('on')
plt.show()
```

ค่า TAC มีนัยสำคัญที่ Lag 12, 24, 36 และลดลงเรื่อยๆ อย่างรวดเร็ว ส่วนค่า TPAC มีนัยสำคัญที่ Lag 12 และสิ้นสุดลง
ดังนั้นแบบจำลอง SARMA ควรเป็น SARMA(1,0)$_{12}$

นอกจากนี้ค่า TAC ที่สดลงอย่างรวดเร็วหลัง Lag 5 และค่า TPAC ที่ลดลงอย่างรวดเร็วหลัง Lag 4 ทำให้แบบจำลองของ $\nu_t$ ควรเป็น ARAM(4,5)

```python
from statsmodels.tsa.arima.model import ARIMA

# Split the data into train and test
train_size = int(len(Profit['Profit']) * 0.8)
train, test = Profit['Profit'][0:train_size], Profit['Profit'][train_size:len(Profit['Profit'])]

# Fit the ARIMA model on the training dataset
model_train = ARIMA(train, order=((0,0,1), 0, (1,0,1,1)), seasonal_order=(1, 0, 0, 12))
model_train_fit = model_train.fit()

print(model_train_fit.summary())

# Forecast on the test dataset
test_forecast = model_train_fit.get_forecast(steps=len(test))
test_forecast_series = pd.Series(test_forecast.predicted_mean)
```

ทำการวิเคราะห์ Residual เพื่อยืนยันว่า $\epsilon_t \mathop{\sim}\limits^{\mathrm{i.i.d.}} \mathcal{N}(0,\sigma^2)$

```python
# line plot of residuals
from pandas import DataFrame
residuals = DataFrame(model_train_fit.resid)
residuals.plot()
plt.axis('tight')
plt.grid('on')
plt.show()
# density plot of residuals
residuals.plot(kind='kde')
plt.axis('tight')
plt.grid('on')
plt.show()
# summary stats of residuals
print(residuals.describe())
```

```python
fig, axes = plt.subplots(2,1, dpi= 100)
plot_acf(residuals, lags=20, ax=axes[0],title="ค่า SAC ของ Residual")
plot_pacf(residuals, lags=20, ax=axes[1],title="ค่า PAC ของ Residual")
plt.axis('tight')
plt.grid('on')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
# ADF Test
adf_test = adfuller(residuals)
print('ADF Statistic: %f' % adf_test[0])
print('p-value: %f' % adf_test[1])

# Ljung-Box Q 
import statsmodels.api as sm
sm.stats.acorr_ljungbox(residuals, lags=[10], return_df=True)
```

```python
# Forecast on the test dataset
test_forecast = model_train_fit.get_forecast(steps=len(test))
test_forecast_series = pd.Series(test_forecast.predicted_mean)

# Create a plot to compare the forecast with the actual test data

mpl.rcParams['font.family'] = 'Tahoma'

plt.style.use('seaborn-v0_8-poster')
plt.figure(figsize=(14,7))
plt.plot(train, label='Training Data')
plt.plot(test, label='Actual Data')
plt.plot(test_forecast_series, label='Forecasted Data', color='green')
plt.fill_between(test.index, 
                 test_forecast.conf_int().iloc[:, 0], 
                 test_forecast.conf_int().iloc[:, 1], 
                 color='k', alpha=.15)
plt.title('ARIMA((0,0,1),0,(1,0,1,1))(1, 0, 0, 12) Model Evaluation')
plt.xlabel('เดือน')
plt.ylabel('กำไร')
plt.legend()
plt.grid('minor')
plt.show()
```

**ตัวอย่าง** พิจารณาข้อมูลมูลค่าการส่งออกของสินค้ารายไตรมาสของบริษัทแห่งหนึ่ง อยู่ในไฟล์ชื่อ ch3_5.xlsx

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import matplotlib as mpl
import matplotlib.pyplot as plt   # data visualization


import pandas as pd 
export = pd.read_excel(r'input/ch3_5.xlsx',sheet_name='Wei')
export.head()
```

```python

mpl.rcParams['font.family'] = 'Tahoma'
plt.style.use('seaborn-v0_8-poster')
plt.plot(export['Export'])
plt.title('ผลกำไรต่อเดือนของบริษัทแห่งหนึ่ง (หมื่นบาท)')
plt.xlabel('เดือน')
plt.ylabel('กำไร')
plt.axis('tight')
plt.grid('minor')
plt.grid('on')
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig, axes = plt.subplots(2,1, dpi= 100)
plot_acf(export['Export'], lags=50, ax=axes[0],title="ค่า SAC ของมูลค่าส่งออก $Export_t$")
plot_pacf(export['Export'], lags=50, ax=axes[1],title="ค่า PAC ของมูลค่าส่งออก, $Export_t$")
plt.axis('tight')
plt.grid('on')
plt.show()
```

จากค่า TAC ที่ลดลงอย่างช้าๆ หมายความว่าอนุกรมเวลาไม่มีความนิ่ง   ใช้ First Differencing เพื่อปรับอนุกรมเวลาให้มีความนิ่ง

```python
DExport = export['Export'].diff().dropna()
DExport.head()
```

```python
plt.style.use('seaborn-v0_8-poster')
plt.plot(DExport)
plt.title('First Difference ของมูลค่าการส่งออก')
plt.xlabel('ลำดับที่')
plt.ylabel('$1^{st}$ ผลต่างของยอดส่งออก')
plt.axis('tight')
plt.grid('minor')
plt.grid('on')
```

ใช้การทดสอบ ADF-Test แล้วยังไม่สามารถปฎิเสธ $H_0$: Unit Root ได้

```python
from statsmodels.tsa.stattools import adfuller
# ADF Test
adf_test = adfuller(DExport)
print('ADF Statistic: %f' % adf_test[0])
print('p-value: %f' % adf_test[1])

# Ljung-Box Q 
import statsmodels.api as sm
sm.stats.acorr_ljungbox(DExport, lags=[10], return_df=True)
```

ตรวจสอบค่า TAC และ TPAC ว่า Seasonal Variation ทำให้อนุกรมเวลาไม่เป็น Stationary หรือไม่

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
fig, axes = plt.subplots(2,1, dpi= 100)
plot_acf(DExport, lags=50, ax=axes[0],title="ค่า SAC ของผลต่างมูลค่าส่งออก $DExport_t$")
plot_pacf(DExport, lags=50, ax=axes[1],title="ค่า PAC ของผลต่างมูลค่าส่งออก, $DExport_t$")
plt.axis('tight')
plt.grid('on')
plt.show()
```

จากค่า TAC พบว่า Lag 4, 8 และ 12 ลดลงช้ามาก ดังนั้นจะใช้ Seasonal Differencing โดย $s=4$ หรือ $\Delta_{s=4}\Delta X_t$ เพื่อทำให้อนุกรมเวลามีความนิ่ง

```python
DSDExport = DExport.diff(4).dropna()# ADF Test
adf_test = adfuller(DSDExport)
print('ADF Statistic: %f' % adf_test[0])
print('p-value: %f' % adf_test[1])

# Ljung-Box Q 
sm.stats.acorr_ljungbox(DSDExport, lags=[10], return_df=True)
```

ADF-Test ยืนยันว่า $\Delta_4 \Delta Export_t$ เป็น Stationary ทำให้พร้อมสร้างแบบจำลองตามแนวคิดของ Box-Jenkins

```python
plt.style.use('seaborn-v0_8-poster')
plt.plot(DSDExport)
plt.title('First Seasonal Differencing  ของมูลค่าการส่งออก')
plt.xlabel('ลำดับที่')
plt.ylabel('$\Delta_4 \Delta Export_t$')
plt.axis('tight')
plt.grid('minor')
plt.grid('on')
```

```python
fig, axes = plt.subplots(2,1, dpi= 100)
plot_acf(DSDExport, lags=50, ax=axes[0],title="ค่า SAC ของค่า $\Delta_4 \Delta Export_t$")
plot_pacf(DSDExport, lags=50, ax=axes[1],title="ค่า PAC ของผลค่า $\Delta_4 \Delta Export_t$")
plt.axis('tight')
plt.grid('on')
plt.show()
```

- พิจารณา SAC ของ Seasonal Variation พบว่าสิ้นสุดหลัง Lag 4  ส่วน Lag 1-3 และ Lag 5 เป็นต้นไปซึ่งเป็นส่วนของ ARMA(p,q) มองว่าสิ้นสุดหลัง Lag ที่ 1
- พิจารณา SPAC ของ Seasonal Variation ก็พบว่าสิ้นสุดหลัง Lag 4 เช่นกัน   ส่วน Lag 1-3 และ Lag 5 นั้นมีการลดตัวอย่างรวดเร็ว

แบบจำลองที่เลือกให้ $\Delta_4\Delta X_t$ คือแบบ $ARMA(0,1)(1,1)_4$  เพราะฉะนั้นแบบจำลอง SARIMA ของยอดการส่งออกคือ
$$ARIMA(p,d,q)(P,D,Q)_s : ARIMA(0,1,1)(1,1,1)_4$$


```python
from statsmodels.tsa.arima.model import ARIMA

# Split the data into train and test
train_size = int(len(export['Export']) * 0.8)
train, test = export['Export'][0:train_size], export['Export'][train_size:len(export['Export'])]

# Fit the ARIMA model on the training dataset
model_train = ARIMA(train, order=(1, 0, 1), seasonal_order=(1, 1, 0, 4),trend='t')
model_train_fit = model_train.fit()

print(model_train_fit.summary())

# Forecast on the test dataset
test_forecast = model_train_fit.get_forecast(steps=len(test))
test_forecast_series = pd.Series(test_forecast.predicted_mean)
```

```python
# line plot of residuals
from pandas import DataFrame
residuals = DataFrame(model_train_fit.resid)
residuals.plot()
plt.axis('tight')
plt.grid('on')
plt.show()
# density plot of residuals
residuals.plot(kind='kde')
plt.axis('tight')
plt.grid('on')
plt.show()
# summary stats of residuals
print(residuals.describe())
```

```python
fig, axes = plt.subplots(2,1, dpi= 100)
plot_acf(residuals, lags=20, ax=axes[0],title="ค่า SAC ของ Residual")
plot_pacf(residuals, lags=20, ax=axes[1],title="ค่า PAC ของ Residual")
plt.axis('tight')
plt.grid('on')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
# ADF Test
adf_test = adfuller(residuals)
print('ADF Statistic: %f' % adf_test[0])
print('p-value: %f' % adf_test[1])

# Ljung-Box Q 
import statsmodels.api as sm
sm.stats.acorr_ljungbox(residuals, lags=[10], return_df=True)
```

```python
# Forecast on the test dataset
test_forecast = model_train_fit.get_forecast(steps=len(test))
test_forecast_series = pd.Series(test_forecast.predicted_mean)

# Create a plot to compare the forecast with the actual test data

mpl.rcParams['font.family'] = 'Tahoma'

plt.style.use('seaborn-v0_8-poster')
plt.figure(figsize=(14,7))
plt.plot(train, label='Training Data')
plt.plot(test, label='Actual Data')
plt.plot(test_forecast_series, label='Forecasted Data', color='green')
plt.fill_between(test.index, 
                 test_forecast.conf_int().iloc[:, 0], 
                 test_forecast.conf_int().iloc[:, 1], 
                 color='k', alpha=.15)
plt.title('ARIMA(1,0,1)(1, 1, 0, 4) Model Evaluation')
plt.xlabel('เดือน')
plt.ylabel('กำไร')
plt.legend()
plt.grid('minor')
plt.show()
```
